```python
import os
import re
import numpy as np
import pandas as pd
```

```python
def collect_accuracy(path, id):
    right = 0
    total = 0
    for i in range(1,4):
        filename = f"{path}/{id}/codesynth-block{i}-final-{id}-1.txt"
        
        try:
            with open(filename, encoding='utf16', errors='ignore') as f:
                lines = f.readlines()
                for line in lines:
                    if re.search('ACC', line):
                        components = line.split()
                        acc = int(components[-1])
                        total += 1
                        right += acc
        except:
            print(f"can't open file {filename}")
    print(id, right, total)
    if total == 0:
        return 0
    return right/total
        
```

```python
# read in data from storage folder
path = "/storage2/fmridata/fmri-data-shapes"
folders = os.listdir(path)

for f in folders:
    if re.match('[0-9]{3}', f):
        id = f
        accuracy = collect_accuracy(path, id)
        print(f"{accuracy}\n")
    # break
        


# iterate through participant directories

# find csv files of task performance - block1, block2, block3

# read as df

# find accuracy column

```
