```python
from nilearn import plotting, surface, datasets
```

```python
results = "../results/nov_int_seed133_tfce_corrp_tstat1.nii.gz"
fsaverage = datasets.fetch_surf_fsaverage()
texture = surface.vol_to_surf(results, fsaverage.pial_left)

plotting.plot_surf_stat_map(fsaverage.infl_left, texture,
                            hemi='left', title='Left hemisphere',
                            threshold=1, cmap='cold_hot', colorbar=True,
                            bg_map=fsaverage.sulc_left)

texture_right = surface.vol_to_surf(results, fsaverage.pial_right)
plotting.plot_surf_stat_map(fsaverage.infl_right, texture_right,
                            hemi='right', title='Right hemisphere',
                            threshold=1, cmap='cold_hot', colorbar=True,
                            bg_map=fsaverage.sulc_right)
```
